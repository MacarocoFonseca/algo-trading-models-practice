````python
import yfinance as yf
import talib as ta
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
````

````python
tesla_data = yf.download('TSLA', '2019-03-19', '2021-02-10')
tesla_data
````

````python
tesla_data['SMA'] = ta.SMA(tesla_data['Adj Close'], 14)
tesla_data
````

````python
tesla_data['SMA'] = pd.to_datetime(tesla_data['SMA'])
tesla_data.index = tesla_data.index.tz_localize('UTC')
````

````python
tesla_data['SMA'].plot()
plt.show()
````

````python
df_tesla = pd.DataFrame(tesla_data)

# Convert the index to datetime
df_tesla.index = pd.to_datetime(df_tesla.index)
df_tesla
````

# Line Graph of close price

````python
# Set the plot style
plt.style.use('seaborn-darkgrid')

# Plotting the close price column
df_tesla['Close'].plot(figsize=(10, 7), color='green')
# Setting the graph title
plt.title("Closing Price of Tesla", fontsize=14)
# Setting the x-axis title
plt.xlabel('Year-Month', fontsize=12)
# Setting the y-axis title
plt.ylabel('Price ($)', fontsize=12)
# Function for showing graph
plt.show()
````

## Scatter Plot of Two Stocks' Close Price

Scatter plot is used to study the relationship between two variables. To plot the scatter plot of two columns in a dataframe, you will use the `scatter` method of `matplotlib`. The syntax is shown below:

Syntax: 
```python
import matplotlib.pyplot as plt
plt.scatter(col1, col2)
```
1. **col1**: First column (x-axis)
2. **col2**: Second column (y-axis)

````python
# Fetch the JP Morgan price data
data = yf.download('jpm', start='2019-03-19', end='2021-02-10')

# Display the data
df_jpm = pd.DataFrame(data)
df_jpm
````

````python
# Plot of Telsa vs JPM
plt.figure(figsize=(10,7))
plt.scatter(df_tesla['Close'], df_jpm['Close'])

# Set the title of the plot
plt.title('Telsa Price Vs. JPM Price', fontsize=14)
# Setting the x-axis title
plt.xlabel('Telsa Price', fontsize=12)
# Setting the y-axis title
plt.ylabel('JPM Price', fontsize=12)
# Function for showing graph
plt.show()
````

## Histogram of Daily Returns

To plot the histogram of a column in a dataframe, you will use the `plot` method of `pandas`. The syntax is shown below:

Syntax: 
```python
DataFrame[col].plot(kind='hist')
```
1. **col**: Column name
2. **kind**: Kind of plot (in case of histogram, it is `'hist'`)

````python
# Histogram of daily returns
plt.figure(figsize=(10,7))
df_jpm['Close'].plot(kind='hist')

# Set the title of the plot
plt.title('close of JPM', fontsize=14)
# Setting the x-axis title
plt.xlabel('Returns', fontsize=12)
# Setting the y-axis title
plt.ylabel('Frequency', fontsize=12)
# Function for showing graph
plt.show()
````

````python
# Data to be stored in the dataframe
stocks_info = [['FB', 50, 330], ['AMZN', 10, 3200], [
    'AAPL', 80, 120], ['NFLX', 40, 500], ['GOOG', 20, 2300]]

# Create a dataframe
stock_data = pd.DataFrame(stocks_info, columns=[
                          'ticker', 'quantity', 'price'], index=['S1', 'S2', 'S3', 'S4', 'S5'])

# Calculate the dollar value
stock_data['dollar_value'] = stock_data['quantity'] * stock_data['price']

# Plot the bar graph
plt.bar(stock_data['ticker'],stock_data['dollar_value'])

# Set title and axes labels
plt.xlabel('Ticker')
plt.ylabel('Dollar Value')
plt.title('Dollar Value of Tickers')
plt.show()
````
